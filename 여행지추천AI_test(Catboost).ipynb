## 여행 정보

```python
df_travel = pd.read_csv('csv\\tn_travel_여행_A.csv')
```

## 여행객 정보

## 데이터 병합(merge)

```python
df = pd.merge(df_place, df_travel, on='TRAVEL_ID', how='left')
df = pd.merge(df, df_traveler, on='TRAVELER_ID', how='left')
```

## 특정 여행 데이터 살펴보기

```python
df[df['TRAVEL_ID'] == 'a_a015688']
```

## 데이터셋 전처리
### 1단계

```python
df_filter = df[~df['TRAVEL_MISSION_CHECK'].isnull()].copy()
df_filter.loc[:, 'TRAVEL_MISSION_INT'] = df_filter['TRAVEL_MISSION_CHECK'].str.split(';').str[0].astype(int)
```

```python
df_filter
```

### 2단계
* 필요한 컬럼만 추출
* 결측치(비어있는 행) 삭제

```python
df_filter.columns
```

```python
df_filter = df_filter[[
    'GENDER','AGE_GRP',
    'TRAVEL_STYL_1', 'TRAVEL_STYL_2',
    'TRAVEL_STYL_3', 'TRAVEL_STYL_4',
    'TRAVEL_STYL_5', 'TRAVEL_STYL_6',
    'TRAVEL_STYL_7', 'TRAVEL_STYL_8',
    'TRAVEL_MOTIVE_1', 'TRAVEL_COMPANIONS_NUM',
    'TRAVEL_MISSION_INT', 'VISIT_AREA_NM',
    'DGSTFN'
]]
```

```python
df_filter = df_filter.dropna()
```

```python
df_filter
```

### 범주형 데이터 변환
* 범주형 데이터는 string 또는 integer 형태여야 함

```python
categorical_features_names = [
    'GENDER',
    'TRAVEL_STYL_1', 'TRAVEL_STYL_2', 'TRAVEL_STYL_3',
    'TRAVEL_STYL_4', 'TRAVEL_STYL_5', 'TRAVEL_STYL_6',
    'TRAVEL_STYL_7', 'TRAVEL_STYL_8',
    'TRAVEL_MOTIVE_1', 'TRAVEL_MISSION_INT', 'VISIT_AREA_NM'
]
```

```python
df_filter[categorical_features_names[1:-1]] = df_filter[categorical_features_names[1:-1]].astype(int)
```

```python
df_filter
```

### 학습/테스트 데이터 분할

```python
from sklearn.model_selection import train_test_split
```

```python
train_data, test_data = train_test_split(df_filter, test_size=0.2, random_state=0)
```

```python
print(train_data.shape)
print(test_data.shape)
```

## CatBoost 모델 훈련
### 데이터 형태 정의

```python
from catboost import CatBoostRegressor, Pool
```

```python
train_pool = Pool(train_data.drop(['DGSTFN'], axis=1),
                  label=train_data['DGSTFN'],
                  cat_features=categorical_features_names)
test_pool = Pool(test_data.drop(['DGSTFN'], axis=1),
                label=test_data['DGSTFN'],
                cat_features=categorical_features_names)
```

### 모델 정의 및 학습

```python
model = CatBoostRegressor(
    loss_function='RMSE',
    eval_metric='MAE',
    task_type='GPU',
    depth=6,
    learning_rate=0.01,
    n_estimators=2000
)
```

```python
model.fit(train_pool, eval_set=test_pool,
          verbose=500, plot=True)
```

## 서비스 구현 시 참고

```python
area_names = df_filter[['VISIT_AREA_NM']].drop_duplicates()
```

```python
traveler = {
    'GENDER': '남',
    'AGE_GRP': 20.0,
    'TRAVEL_STYL_1': 1,
    'TRAVEL_STYL_2': 1,
    'TRAVEL_STYL_3': 2,
    'TRAVEL_STYL_4': 3,
    'TRAVEL_STYL_5': 2,
    'TRAVEL_STYL_6': 2,
    'TRAVEL_STYL_7': 2,
    'TRAVEL_STYL_8': 3,
    'TRAVEL_MOTIVE_1': 8,
    'TRAVEL_COMPANIONS_NUM': 0.0,
    'TRAVEL_MISSION_INT': 3,
}
```

```python
results = pd.DataFrame([], columns=['AREA', 'SCORE'])
```

```python
for area in area_names['VISIT_AREA_NM']:
    input = list(traveler.values())
    input.append(area)

    score = model.predict(input)

    results = pd.concat([results, pd.DataFrame([[area, score]], columns=['AREA', 'SCORE'])])

results.sort_values('SCORE', ascending=False)[:10]
```
